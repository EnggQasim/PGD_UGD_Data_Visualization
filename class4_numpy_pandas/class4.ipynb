### np.fromfunction

```python
def f(x, y):
    print("x",x)
    print("y",y)
    return x + y

b = np.fromfunction(f, (5, 4), dtype=int)
b
```

# Shallow Copy

```python
a = np.array([1,2,3])
b = a # nickname
c = b # nickname

print(a)
print(b)
print(c)
```

```python
c[0] = 20# 
print(a)
print(b)
print(c)
```

## Deep copy

```python
a = np.array([1,2,3])
b = a.copy()
c = b.copy()

a[0] = 20# 
print(a)
print(b)
print(c)
```
